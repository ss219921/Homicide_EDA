# Homicide

```python
import os
import pandas as pd
import numpy as np

crime=pd.read_csv('D:/Shweta/EDA/Files/Homicide.csv', low_memory=False)
```

```python
crime.head()
```

```python
# 1. Number of incidents per States per year

incidents_per_state = crime.groupby(['State','Year'])['Record ID'].count().nlargest().reset_index()
incidents_per_state = incidents_per_state.rename(columns = {'Record ID':'Total Incident'})
incidents_per_state.sort_values(by ='Total Incident', ascending = False)
incidents_per_state.head()
```

```python
# 2. Which month wife's have more tendency to hurt husband?

husband =crime[(crime['Victim Sex'] == 'Male') & (crime['Perpetrator Sex'] == 'Female') & (crime['Relationship'] == 'Husband')]
month = husband.groupby('Month')['Record ID'].count().reset_index()
month = month.rename(columns = {'Record ID': 'Total Incident'})
month.sort_values(by = 'Total Incident', ascending = False)
```

```python
# 3. Number of incident involving Ex-husband vs Ex wife (older or yonger)

incident = crime[['Relationship','Record ID','Victim Sex','Perpetrator Sex']]
incident = incident[((incident['Relationship'] == 'Ex-Wife') &
                                    (incident['Victim Sex']=='Female') & 
                                     (incident['Perpetrator Sex']=='Male')) | 
                                    ((incident['Relationship'] == 'Ex-Husband') &
                                     (incident['Victim Sex']=='Male') & 
                                     (incident['Perpetrator Sex']=='Female'))]
incident = incident.groupby('Relationship')['Record ID'].count().reset_index()
incident = incident.sort_values(by = 'Record ID',ascending = False)
incident
```

```python
# 4. Number of incident between Employee and employer?

Emp = crime[crime['Relationship'] == 'Employee'] | crime[crime['Relationship'] == 'Employer']
len(Emp)
```

```python
# 5. Number of incident based on family relationship

crime['Family'] = 0
for i in range(len(crime)):
    if (crime.loc[i,'Relationship'] == "Brother") & (crime.loc[i, 'Perpetrator Sex'] == "Male"):
        crime.loc[i,'Family'] = "Brother Killed by Brother"
    elif (crime.loc[i, 'Relationship'] == "Brother") & (crime.loc[i, 'Perpetrator Sex'] == "Female"):
        crime.loc[i, 'Family'] = "Brother Killed by Sister"
    elif (crime.loc[i, 'Relationship'] == "Sister") & (crime.loc[i, 'Perpetrator Sex'] == "Female"):
        crime.loc[i, 'Family'] = "Sister Killed by Sister"
    elif (crime.loc[i, 'Relationship'] == "Sister") & (crime.loc[i, 'Perpetrator Sex'] == "Male"):
        crime.loc[i, 'Family'] = "Sister Killed by Brother"
    elif (crime.loc[i, 'Relationship'] == "Mother") & (crime.loc[i, 'Perpetrator Sex'] == "Male"):
        crime.loc[i,'Family'] = "Mother Killed by Son"
    elif (crime.loc[i, 'Relationship'] == "Mother") & (crime.loc[i, 'Perpetrator Sex'] == "Female"):
        crime.loc[i, 'Family'] = "Mother Killed by Daughter"
    elif (crime.loc[i, 'Relationship'] == "Son") & (crime.loc[i, 'Perpetrator Sex'] == "Female"):
        crime.loc[i, 'Family'] = "Son Killed by Mother"
    elif (crime.loc [i, 'Relationship'] == "Son") & (crime.loc[i, 'Perpetrator Sex'] == "Male"):
        crime.loc[i, 'Family'] = "Son Killed by Father"
    elif (crime.loc[i, 'Relationship'] == "Daughter") & (crime.loc[i, 'Perpetrator Sex'] =="Female"):
        crime.loc[i, 'Family'] = "Daughter Killed by Mother"
    elif (crime.loc[i, 'Relationship'] == "Daughter") & (crime.loc[i, 'Perpetrator Sex'] =="Male"):
        crime.loc[i, 'Family'] = "Daughter Killed by Father"
    elif (crime.loc[i, 'Relationship'] == "Wife") & (crime.loc[i, 'Perpetrator Sex'] =="Male"):
        crime.loc[i, 'Family'] = "Wife Killed by Husband"
    elif (crime.loc[i, 'Relationship'] == "Husband") & (crime.loc[i, 'Perpetrator Sex'] =="Female"):
        crime.loc[i, 'Family'] = "Husband Killed by wife"
    elif (crime.loc[i, 'Relationship'] == "Father") & (crime.loc[i, 'Perpetrator Sex'] == "Female"):
        crime.loc[i, 'Family'] = "Father Killed by Daughter"
    elif (crime.loc[i, 'Relationship'] == "Father") & (crime.loc[i, 'Perpetrator Sex'] == "male"):
        crime.loc[i, 'Family'] = "Father killed by Son"
                    
crime


```
